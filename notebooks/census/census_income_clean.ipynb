# clean census data

```python
import pandas as pd
import numpy as np
import os
import re
```

```python
csv_path = os.path.join('..', '..', 'raw', 'census', 'income_by_zip', 'ACS_16_5YR_S1901_with_ann.csv')
df = pd.read_csv(csv_path)
df.head()
```

```python
df.rename(columns=df.loc[0], inplace=True)
df.drop(0, inplace=True)
df.set_index('Id2', inplace=True)
df.index.name = 'zip_code'
df.head()
```

```python
df.columns
```

```python
mean = df.filter(like='Mean')
median = df.filter(like='Median')
mean_med = mean.merge(median, left_index=True, right_index=True)
mean_med.columns
```

```python
mean_med.columns = mean_med.columns.map(lambda x: re.sub('Households; |Estimate; | \(dollars\)', '', x))
mean_med.columns = mean_med.columns.map(lambda x: re.sub('Margin of Error;', 'MoE', x))
mean_med.head()
```

```python
mean_med.replace(to_replace='-', value=np.NaN, inplace=True)
mean_med.replace(to_replace='N', value=np.NaN, inplace=True)
mean_med.replace(to_replace='**', value=np.NaN, inplace=True)
mean_med.dropna(inplace=True)
```

```python
mean_path = os.path.join('..', '..', 'data', 'census_income_by_zip.csv')
mean_med.to_csv(mean_path)
```
