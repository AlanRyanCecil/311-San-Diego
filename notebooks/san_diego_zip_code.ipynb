## Zip Code Data Cleaning

```python
#Dependencies
import os
import pandas as pd
import numpy as np
```

```python
# File to Read and Write
file_path = os.path.join('zip_codes_states.csv')
output_path = os.path.join('san_diego_zip_code.csv')
zip_code_data = pd.read_csv(file_path)
zip_code_data.head()
```

## City of San Diego Zip Codes

```python
# find all the zip codes for San Deigo by loc
san_diego_data = zip_code_data.loc[zip_code_data['city']=='San Diego',:]
san_diego_data .head()
```

```python
san_diego_data['state'].value_counts()
```

```python
# drop San Diego in TX
san_diego_data = san_diego_data.loc[san_diego_data['state']=='CA',:]
san_diego_data.head()
```

```python
# Reset Index
san_diego_data = san_diego_data.reset_index(drop=True)
san_diego_data.head()
```

```python
output_file = san_diego_data.to_csv(output_path)
```
